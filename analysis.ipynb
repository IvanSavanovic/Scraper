```python
stopwords = [l[:-1] for l in open("stopwords-hr.txt", encoding='utf-8').readlines()]
keywords_by_month = collections.OrderedDict()
charaters = list(string.punctuation) + ['–']
for month, posts in post_by_month.items():
    keywords = []
    #month = date.split("/")[1]
    for post in posts:
        text = post.lower()
        for c in charaters:
            text = text.replace(c, "")
        text = "".join([c for c in text if not c.isdigit()])
        text = text.split()
        text = [w for w in text if w not in stopwords]
        keywords.extend(text)
    keywords_by_month[int(month)] = nltk.FreqDist(keywords)
    keywords = []
print(keywords_by_month)
```

```python
for month, fd in keywords_by_month.items():
    keywords_by_month[month] = sorted(fd.items(), reverse=True, key = lambda x: x[1]) 
print(keywords_by_month[11])
```

```python
Jaccard_max = np.empty((11, 11), dtype = np.float32)
keyword_item = sorted(keywords_by_month.items(), key = lambda x: [0])

for i, (month, words) in enumerate(keyword_item):
    topx = set([w[0] for w in words[:25]]) 
    for j, (month1, words1) in enumerate(keyword_item):
       topy = set([w[0] for w in words1[:25]])
       Jaccard_max[i][j] = len((topx & topy))/len((topx | topy))
#print(Jaccard_max)
```

```python
for row in range(11):
    for col in range(11):
        print("{:.2}".format(Jaccard_max[row, col]), end=" ")
    print()
```
